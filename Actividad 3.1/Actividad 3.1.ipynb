```python
#Importamos las librerias
import pandas as pd 
import numpy as np 
import matplotlib as plt 
```

```python
#Actividad 3.1 
df = pd.read_csv('Ventas_totales.csv')
df
```

```python
df.isnull().sum()
#Con valores nulos: 
#salon_ventas, tarjetas_debito, tarjetas_credito, otros_medios, subtotal_ventas_alimentos_bebidas, bebidas 
#almacen, panaderia, lacteos, carnes, verduleria_fruteria, alimentos_preparados_rotiseria, 
#indumentaria_calzado_textiles_hogar, electronicos_articulos_hogar, otros
```

```python
#Se crea una copia del archivo para poder rellenar los datos 
df2=df.copy()
df2.info()
df2.describe()
```

```python
#Se rellenan los datos 

#Llenado por media 
df2['salon_ventas']=df2['salon_ventas'].fillna(round(df['salon_ventas'].mean(),1))
df2['tarjetas_debito']=df2['tarjetas_debito'].fillna(round(df['tarjetas_debito'].mean(),1))
df2['tarjetas_credito']=df2['tarjetas_credito'].fillna(round(df['tarjetas_credito'].mean(),1))
df2['otros_medios']=df2['otros_medios'].fillna(round(df['otros_medios'].mean(),1))
df2['subtotal_ventas_alimentos_bebidas']=df2['subtotal_ventas_alimentos_bebidas'
                                             ].fillna(round(df['subtotal_ventas_alimentos_bebidas'].mean(),1))
#Llenado por ffill
df2['bebidas']=df2['bebidas'].fillna(method="ffill")
df2['almacen']=df2['almacen'].fillna(method="ffill")
df2['panaderia']=df2['panaderia'].fillna(method="ffill")
df2['lacteos']=df2['lacteos'].fillna(method="ffill")

#Llenado por bfill
df2['carnes']=df2['carnes'].fillna(method="bfill")
df2['verduleria_fruteria']=df2['verduleria_fruteria'].fillna(method="bfill")
df2['alimentos_preparados_rotiseria']=df2['alimentos_preparados_rotiseria'
                                          ].fillna(method="bfill")

#Llenado por median 
df2['indumentaria_calzado_textiles_hogar']=df2['indumentaria_calzado_textiles_hogar'
                                               ].fillna(round(df['indumentaria_calzado_textiles_hogar'].median(),1))
df2['electronicos_articulos_hogar']=df2['electronicos_articulos_hogar'
                                        ].fillna(round(df['electronicos_articulos_hogar'].median(),1))
df2['otros']=df2['otros'].fillna(round(df['otros'].median(),1))
```

```python
#Comprobamos que ya no existan valores nulos 
valores_nulos = df2.isnull().sum()
valores_nulos
```

**JUSTIFICACION**

El código utiliza diferentes métodos de rellenado para manejar valores nulos en un DataFrame, seleccionando la técnica más adecuada según la naturaleza de cada columna. Para mantener la tendencia central de los datos, se emplea el promedio redondeado en variables de transacciones y ventas, como salon_ventas y tarjetas_debito. La propagación hacia adelante (ffill) se utiliza en columnas donde los valores faltantes pueden seguir una secuencia lógica, como bebidas y almacen; en otras, como carnes y verduleria_fruteria, se utiliza la propagación hacia atrás (bfill). Para evitar la influencia de valores inusuales, se utiliza la mediana redondeada en columnas como ropa_calzado_textiles_hogar, donde la distribución puede ser sesgada. Estos métodos permiten que los datos sean rellenados de manera adecuada y coherente, manteniendo la integridad del análisis.
